Introduction:

Minority groups in the US have been severely harmed by systemic gaps in healthcare access, especially when it comes to chronic illnesses like type 2 diabetes. Minority populations encounter a multitude of obstacles, such as restricted availability of healthcare services and elevated incidence of underinsurance or uninsurance as a result of working in low-wage industries. As a result, among minorities, diseases like diabetes, hypertension, and cancer are frequently detected at a later stage, impeding early management and aggravating health disparities. Minority populations have disproportionately high rates of illness and mortality when compared to non-minority groups, which is the result of delayed diagnosis and treatment that perpetuates cycles of poor health outcomes.

Addressing these gaps requires an understanding of the intricate interactions that exist between chronic illnesses, healthcare access, and race. Promoting fair healthcare practices and regulations is crucial to reducing the structural obstacles that impede minority populations' access to high-quality medical treatment. Through an examination of the relationship between race, access to healthcare, and the occurrence of long-term illnesses such as type 2 diabetes, we may illuminate the extent of these differences and endeavor to advance health equity for all groups.

In this project, we are going to take a deeper dive into whether there really is a coorleation between race and health and how large of a role it plays. We will be analyzing multiple factors that can play a part, and look into the data to make predictions about the potential damage this can do to minority communities in the United States.

```python
dataset= pd.read_excel("health3a_cycle03.xlsx", header=4)
```

As our project's main focus is on race and how it plays a part in predicting type 2 diabetes, we only wanted to analyze the characteristic of Origin and Race. The data set gives us 5 groups of races to look at: Hispanic or Latino, White, Black, Asian, and someone who is bi-racial. We are able to look at multiple data points for each racial group. The first is the sum of the total number of people insured, the second is how many of those people have public insurance, the third is number of people who have private insurance, and then we have the the number of people who are not insured.

This dataset contained a lot of characterstics that we didn't deem necessary for our project, such as Age, Sexual Orientation, and Gender. As a reselt, we had to manipulate our dataframe to only get information we needed. We started by removing the bottom 2 rows, which were rows of NaN used as filler rows in the data. We also had to drop the first row because the data table used 2 rows for the column headers. 

To actually remove the characteristics we didn't need, we saw that was between the first 29, and the last 90, so we removed those from the dataset using iloc. After resetting the index again, we had all the correct information we needed. 

Another issue that we had to manipulate the table for was that the insurance was labeled in a way that would be unhelpful for someone who didn't see the key on the original spreadsheet. Total insurace was labeled Insured*, public insurance was labeled Insured*.1, and private insurance was labeled Insured*.2. 

```python
# Remove the last 2 rows
dataset = dataset.iloc[:-2]

# Reset the index after removing rows
dataset.reset_index(drop=True, inplace=True)
dataset.drop(1, inplace=True)
dataset= dataset.iloc[29:]
dataset= dataset.iloc[:-90]
dataset.reset_index(drop=True, inplace= True)
dataset.rename(columns={'Insured*': 'Total Insured', 'Insured*.1': 'Public Insurance', 'Insured*.2': 'Private Insurance'}, inplace=True)
pd.set_option('display.max_rows', None)
dataset.fillna('', inplace=True)

dataset

#Insured*= Total Insured
#Insured*.1= Public Insurance
#Insured*.2= Private Insurance
```

A helpful graphic for this dataset would be a bargraph of the percent of uninsured vs the total number of insured and uninsured so we could get a better idea of how each racial group stacked up against each other and who was the least and most insured. 

Our prediction during the project was that groups that had a higher percentage of people uninsured were more likely to have Type 2 Diabetes because they don't have the ability to get it checked early. In an article by NCBI (https://www.ncbi.nlm.nih.gov/pmc/articles/PMC7369944/), they had concluded thru their own surveys that "having health insurance was associated with decreased odds for undiagnosed prediabetes and type 2 diabetes in American adults." 

We calculated our Uninsured rate by first converting the dataset into numerics, and then getting the rate by dividing the number of uninsured by the total number of people insured and uninsured. 

```python
dataset['Total Insured'] = pd.to_numeric(dataset['Total Insured'], errors='coerce')
dataset['Uninsured'] = pd.to_numeric(dataset['Uninsured'], errors='coerce')

dataset['Uninsured Rate'] = dataset['Uninsured'] / (dataset['Total Insured']+dataset['Uninsured'])
plt.figure(figsize=(10, 6))
plt.bar(dataset['Select characteristics'], dataset['Uninsured Rate'], color='skyblue')
plt.title('Uninsured Rate by Race')
plt.xlabel('Race')
plt.ylabel('Uninsured Rate')
plt.xticks(rotation=45)
plt.show()
```

```python
cumulative_table['Total'] = cumulative_table.sum(axis=1)

# Calculate the percentage of each race for each state
cumulative_table['PercentWhite'] = cumulative_table['TractWhite'] / cumulative_table['Total'] * 100
cumulative_table['PercentBlack'] = cumulative_table['TractBlack'] / cumulative_table['Total'] * 100
cumulative_table['PercentAsian'] = cumulative_table['TractAsian'] / cumulative_table['Total'] * 100
cumulative_table['PercentHispanic'] = cumulative_table['TractHispanic'] / cumulative_table['Total'] * 100

# Plotting pie charts for each state
states = cumulative_table.index
num_states = len(states)
num_cols = 8  # Number of columns in the subplot grid
num_rows = -(-num_states // num_cols)  # Ceiling division to get the number of rows

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, 3 * num_rows))

for i, state in enumerate(states):
    row = i // num_cols
    col = i % num_cols
    labels = ['White', 'Black', 'Asian', 'Hispanic']
    sizes = [cumulative_table.loc[state, 'PercentWhite'], cumulative_table.loc[state, 'PercentBlack'],
             cumulative_table.loc[state, 'PercentAsian'], cumulative_table.loc[state, 'PercentHispanic']]
    axes[row, col].pie(sizes, labels=None, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 8})
    axes[row, col].set_title(f'{state}', fontsize=10)
    axes[row, col].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Remove any unused subplots
for i in range(num_states, num_rows * num_cols):
    fig.delaxes(axes.flatten()[i])

# Create a legend for the entire figure
legend_labels = ['White', 'Black', 'Asian', 'Hispanic']
legend_colors = ['blue', 'orange', 'green', 'red']
fig.legend(legend_labels, loc='center', bbox_to_anchor=(0.5, 0.05), title='Race', fontsize='small', title_fontsize='small', markerscale=0.5, facecolor='lightgrey')

# Adjust layout to prevent overlap and reduce spacing
plt.subplots_adjust(wspace=0.2, hspace=0.2)

# Show the plot
plt.show()
```

The dataframe below is data pulled from the CDC on estimated crude prevalence of diagnosed diabetes, undiagnosed diabetes, and total diabetes among adults aged 18 years or older, United States.
```python
r = requests.get('https://www.cdc.gov/diabetes/php/data-research/?CDC_AAref_Val=https://www.cdc.gov/diabetes/data/statistics-report/index.html')
data = r.text


# Parse the HTML content
soup = bs(data, 'html.parser')
# Find the table you're interested in, by an ID or the specific attributes
table = soup.find('div', {'class': 'table-responsive'})
rows = table.find_all('tr')
table_data = []
for row in rows:
    cols = row.find_all(['td','th'])
    cols = [ele.text.strip() for ele in cols]
    table_data.append(cols)

#Create a DataFrame from the list of rows
ethnic_df = pd.DataFrame(table_data)
ethnic_df

 
```

We would like to primarly focus on the race and how diabetes is distributed based on race. Below I work on isolating the total diabetes for four the four races, White, Black, Asian, and Hispanic. This dataset isolates the four races we would like to use later for our pie chart to better visiulaize the distribution. 

```python
percentage_arr = []
sum = 0
percentage_arr.append(ethnic_df.iloc[10,3])
percentage_arr.append(ethnic_df.iloc[11,3])
percentage_arr.append(ethnic_df.iloc[12,3])
percentage_arr.append(ethnic_df.iloc[13,3])

percentage_arr = [float(v.split()[0]) for v in percentage_arr]

for val in percentage_arr:
    sum += val

other_race = 100 - sum

percentage_arr.append(other_race)

labels = ['White, non-Hispanic', 'Black, non-Hispanic', 'Asian, non-Hispanic', 'Hispanic','Other']
colors = ['blue', 'orange', 'green', 'red','grey']
plt.pie(percentage_arr, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Total Diabetes Percentage by Race and Ethnicity')
plt.show()
```

The data below is taken from a dataset provided by Peter G Peterson Foundation which gives data on income and wealth in the United States for 2021 and 2022 based on a number of factors. The table below is data just on the four races, Asian, Hispanic, Black, and White and their median yearly income.  

```python
r = requests.get('https://www.pgpf.org/blog/2024/05/income-and-wealth-in-the-united-states-an-overview-of-recent-data')
data = r.text
soup = bs(data, 'html.parser')
# Find the table you're interested in, by an ID or the specific attributes
table = soup.findAll('figure', {'class': 'chart'})

table_count = 0
main_table = table
for tables in table:
    if table_count == 2:
        main_table = tables
        break
    else:
        table_count+=1

rows = main_table.find_all('tr')


table_data = []

for row in rows:
    cols = row.find_all(['td', 'th'])
    cols = [ele.text.strip() for ele in cols]
    table_data.append(cols)

#Create a DataFrame from the list of rows

income_df = pd.DataFrame(data=table_data[1:], columns=table_data[0])

#Display the DataFrame
income_df
```

```python
race_ethnicity = ['Asian', 'White, not Hispanic', 'Hispanic', 'Black']

asian_1 = income_df.iloc[1,1].replace("$", "").replace(",", "")
white_1 = income_df.iloc[2,1].replace("$", "").replace(",", "")
hispanic_1 = income_df.iloc[3,1].replace("$", "").replace(",", "")
black_1 = income_df.iloc[4,1].replace("$", "").replace(",", "")
asian_2 = income_df.iloc[1,2].replace("$", "").replace(",", "")
white_2 = income_df.iloc[2,2].replace("$", "").replace(",", "")
hispanic_2 = income_df.iloc[3,2].replace("$", "").replace(",", "")
black_2 = income_df.iloc[4,2].replace("$", "").replace(",", "")


income_2021 = [float(asian_1), float(white_1), float(hispanic_1), float(black_1)]  # 2021 Median Household Income
income_2022 = [float(asian_2), float(white_2), float(hispanic_2), float(black_2)]  # 2022 Median Household Income

plt.figure(figsize=(8, 6))
plt.bar(race_ethnicity, income_2021, color='blue')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Median Household Income ($)')
plt.title('Median Household Income by Race/Ethnicity for 2021')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()

# Graph for 2022
plt.figure(figsize=(8, 6))
plt.bar(race_ethnicity, income_2022, color='green')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Median Household Income ($)')
plt.title('Median Household Income by Race/Ethnicity for 2022')
plt.xticks(rotation=45)  # Rotate labels for better readability
plt.show()
```
